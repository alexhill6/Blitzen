```python
%%sql
--Basic institution information
CREATE TABLE institution_info (
    UNITID SERIAL PRIMARY KEY,
    ACCREDAGENCY TEXT,
    PREDDEG TEXT,
    HIGHDEG TEXT,
    CONTROL INTEGER,
    REGION TEXT,
    CCBASIC TEXT
);

--Institution financial data
CREATE TABLE institution_financial (
    UNITID SERIAL PRIMARY KEY,
    YEAR INTEGER,
    TUITIONFEE_IN MONEY,
    TUITIONFEE_OUT MONEY,
    TUITIONFEE_PROG MONEY,
    TUITFTE MONEY,
    AVGFACSAL MONEY,
    CDR2 FLOAT,
    CDR3 FLOAT
);

--Institution admissions data
CREATE TABLE institution_admissions (
    UNITID SERIAL PRIMARY KEY,
    YEAR INTEGER,
    ADM_RATE FLOAT,
    SATVR25 FLOAT,
    SATVR75 FLOAT,
    SATMT25 FLOAT,
    SATMY75 FLOAT,
    SATWR25 FLOAT,
    SATWR75 FLOAT,
    SATVRMID FLOAT,
    SATMTMID FLOAT,
    SATWRMID FLOAT,
    ACTCM25 FLOAT,
    ACTCM75 FLOAT,
    ACTEN25 FLOAT,
    ACTEN75 FLOAT,
    ACTMT25 FLOAT,
    ACTMT75 FLOAT,
    ACTWR25 FLOAT,
    ACTWR75 FLOAT,
    ACTCMMID FLOAT,
    ACTENMID FLOAT,
    ACTMTMID FLOAT,
    ACTWRMID FLOAT,
    SAT_AVG FLOAT 
);

--Institution demographic completion rate 
CREATE TABLE institution_completion (
    UNITID SERIAL PRIMARY KEY,
    YEAR INTEGER, 
    C150_4 FLOAT,
    C150_4_WHITE FLOAT,
    C150_4_BLACK FLOAT,
    C150_4_HISP FLOAT,
    C150_4_ASIAN FLOAT,
    C150_4_AIAN FLOAT,
    C150_4_NHPI FLOAT,
    C150_4_2MOR FLOAT,
    C150_4_NRA FLOAT,
    C150_4_UNKN FLOAT,
);

--Institution earnings
CREATE TABLE institution_earnings (
    UNITID SERIAL PRIMARY KEY,
    YEAR INTEGER, 
    MN_EARN_WNE_INC1_P6 FLOAT,
    MN_EARN_WNE_INC2_P6 FLOAT,
    MN_EARN_WNE_INC3_P6 FLOAT,
    MN_EARN_WNE_INDEP0_INC1_P6 FLOAT,
    MN_EARN_WNE_INDEP0_P6 FLOAT,
    MN_EARN_WNE_INDEP1_P6 FLOAT,
    MN_EARN_WNE_MALE0_P6 FLOAT,
    MN_EARN_WNE MALE1_P6 FLOAT
);
```

**Table descriptions:**

- `institution_info`: Contains factual, institution-specific information such as unique identifiers, names, locations, and other basic descriptive attributes related to each institution.
- `institution_financial`: Contains tuition costs and financial information about the institution itself.
- `institution_admissions`: Contains admissions statistics (e.g., acceptance rates, enrollment figures) and standardized test performance, including the 75th percentile SAT or ACT scores for admitted students.
- `institution_completion`: Contains data on institutional completion or graduation rates.
- `institution_earnings`: Contains information on the average post-graduation earnings of students from each institution, tracked across multiple years.


**Design decisions:**

We decided to put the static information about the institution in a table while all the info that changes over time, like completion rate, testing scores, tuition, and financials are in other tables that include Year. 

We then separated the tables into groups of variables, one group being testing scores, one group being average earnings, one group being financial, and the last group being completion rates, considering the goal of producing an analysis report.
